## Lambda Functions <br>
<p>Lambda functions... or "Anonymous Functions" are referring to inline functions with no name. The keyword lambda denotes the no name function, and executes within a single line. Without saving it to a variable; however, it is not able to be used, unless passed in either as a paramater or within list comprehension.<br>Written as "(keyword lambda) (one or more inputs) (colon) (function to be executed)"</p>

#### Syntax

```python
def addTwo(x):
    return x + 2

print(addTwo(8))

# Lambda function syntax
# Write and call the lambda

print((lambda x: x +2)(8)) #lambdas are quick functions, left side variable: right side is what you're returning
```

#### Saving to a Variable

```python
add_two = lambda x: x +2

add_two(9)
```

#### Multiple Inputs

```python
# multiple inputs (not stored or saved to a variable)

print((lambda x, y, z: x*y*z)(2,5,10))

mulitply = lambda x, y, z: x*y*z

multiply(2,5,10)
```

#### Passing a Lambda into a Function

```python
def square(f, num):
    """
        f expects a lamda
        num expects an integer
    """
    
    return f(num)

square(lambda x: x * x,4)
```

#### Returning a Lambda from a Function

```python
# Regular defined function
def multiply_test(num):
    return num * 4

# Function written inside other function

def returnFunc():
    test = 4
    def multiply(num):
        return num * test
    return multiply

f_return = returnFunc()
print(f_return)
print(f_return(25))

# Lambda Returned from regular func

def returnLamb(b,c):
    return lambda x,y: (b+c) / (x+y)

r_lamb = returnLamb(5,7) #essentially saying, the lambda inside will be hardcoded as whatever input you placed and ready to use for another function

print(r_lamb(20,4))
```

#### If Statements within Lambdas

```python
# for order of priority Truthy comes first, if statement, elifs, then Falsey last

# lambda x: (Truthy) if (condition) else (Falsey)

# you can nest "if" conditions inside the lambda

# lambda x: (Truthy variable) if (true condition) else{begining of elif} (Truthy variable) if (true condition) else (Falsey)...

f_condition = lambda num: num * 2 if num > 10 else "The num is 10" if num == 10 else num +2
# if you want an elif statment, it comes after the "else" of the Truthy as a new logic, then ends with another else with a Falsey statment
print(f_condition(8))
print(f_condition(12))
print(f_condition(10))
```

## Map <br>
<p>The map function allows you to iterate over an entire list while running a function on each item of the list. This is why the map function works well with lambda's, because it simplifies things and you write less lines of code.<br>The syntax for a map function is "map(function to be used, list to be used)"<br>However, you must be careful, as the map function returns a map object, not a list. To turn it into a list we use the list() type conversion.</p>

#### Syntax

```python
# runs a function over every iterable (runs a for loop in the array)
# it returns a map object, you must turn it into a list or any iterable type afterwards

# map(func, iterable --(list, dict, tuple, set ))
# normally this map happens with a predefined function -- but lambdas are usable too

def squared(num1, num2):
    if num1 < 10 and num2 <10:
        return num1 **2, num2 **2
    else:
        return num1, num2
    
nums = [4,11,20,8,15,90] #6 values
more_nums = [4,10,3,2,6] #5 values

squared_nums_map = list(map(squared, nums, more_nums))


print(squared_nums_map)
```

```python
def cubed(num):
    if num > 10:
        return num
    else:
        return num ** 3

cube_map = list(map(cubed, nums))
print(cube_map)
```

#### Using Lambda's with Map

```python
# map( (lambda x,y: x+y), list)


squared_nums_lamb = list(map((lambda x,y: (x * x, y * y) if x < 10 and y <10 else (x,y)), nums, more_nums))

print(squared_nums_lamb)
```

## Filter() <br>
<p>Filter's are similar to the map function, where you're able to pass a function argument and a list argument and filter out something from the list based on the conditions passed. Similar to the map function, it returns a filter object, so you need to type convert it to a list()</p>

#### Syntax

```python
# whatever satisfies the condition of the function to True; will be returned


names = ["Aaron","Sandy","Angelica","Jason","Andy"]

def a_names(name):
    if name[0].lower() == "a":
        return True
    else:
        return False

filtered_name_list = list(filter(a_names, names))

print(filtered_name_list)
```

```python
print(list(filter(None,["",10,0,"string"])))

# true stays, false goes. 
# "" and 0 are Falsey
```

#### Using Lambda's with Filter()

```python
a_names_list = list(filter(lambda name: True if name[0].lower() == 'a' else False, name))
```

## Reduce() <br>
<p>Be very careful when using this function, as of Python 3 it's been moved to the 'functools' library and no longer is a built-in function.<br>The creator of Python himself, says to just use a for loop instead.</p>

#### Syntax

```python
# will take a whole iterable, and reduce it down to one value by applying a function to each value and continuing until the end of the array, replacin gther old values
# For example, reduce(lambda x, y: x+y, [1, 2, 3, 4, 5]) calculates ((((1+2)+3)+4)+5)
# need to import from functools

from functools import reduce

alist = [2,4,6,8,10]

def addNums(accumulator, num2):
    print("accumulator: ", accumulator)
    return accumulator + num2

result = reduce(addNums, alist)
print("Result: ", result)
```

#### Using Lambda's with Reduce()

```python
result_lamb = reduce(lambda x,y: x + y, alist)

print(result_lamb)
```

## Recursion <br>
<p>Recursion means that a function is calling itself, so it contanstly executes until a base case is reached. It will then push the returning values back up the chain until the function is complete. A prime example of recursion is computing factorials... such that 5! (factorial) is 5*4*3*2*1 which equals 120.</p>

#### Implementing a Base Case

```python
# in the definition of a function, the same function is used
# will continue running until the base case is reached and will then return values from the base case on up

# Base Case is a conditional statement; it must return True for the repetition to complete

def addNums(num):
    #First, set a Base Case for recursion to stop
    if num <= 1:
        print("addNums(1) = 1")
        return 1
    
    #Recursive call
    else:
        print(f"addNums({num}) = {num} + addNums({num - 1})")
        return num + addNums(num - 1)

#once you get the value for a call, python goes backwards returning the values of the calls that were missed in the loop to solve the call of thr value you are asking for
addNums(5)
```

#### Writing a Factorial Function

```python
#5! = 5*4*3*2*1

def factorial(num):
    if num <= 1:
        return 1
    else:
        return num * factorial(num - 1)

factorial(5)

# iteratively
def factorial_iteration(num):
    answer = 1
    for i in range(1, num+1):
        answer *= i
    return answer

print(factorial_iteration(5))
```

## Generators <br>
<p>Generators are a type of iterable, like lists or tuples. They do not allow indexing, but they can still be iterated through with for loops. They are created using functions and the yield statement.</p>

#### Yield Keyword <br>
<p>The yield keyword denotes a generator, it doesn't return so it won't leave the function and reset all variables in the function scope, instead it yields the number back to the caller.</p>

```python
def my_range(stop, start, step=3):
    while start < stop:
        yield start #Yield denotes a generator
        start += step
        
print(my_range(20,0))

for i in my_range(20,0):
    print(i)
```

<p>Create a generator that takes a number argument and yields that number squared, then prints each number squared until zero is reached.</p>

```python

def generator_range(start, stop, step = -1):
    while start >= 0:
        yield start ** 2
        start += step

for i in generator_range(5,0):
    print(i)
```

`

```python
# F = (9/5)*C + 32
places = [('Nashua',32),("Boston",12),("Los Angelos",44),("Miami",29)]

farhenheit_places = list(map((lambda tup: (tup[0], tup[1]*(9/5) + 32)), places))

# lambda definition: inside the tup of places, return two values, one itself, the other placed in the equation
print(farhenheit_places)
```
